## Data Source Selection

The calendar dataset was initially considered as the source of the target variable. However, `calendar.csv` does not include `price` or `adjusted_price` columns at all — it only contains `listing_id`, `date`, `available`, `minimum_nights`, and `maximum_nights`.

Therefore, the listings dataset was selected as the source of the target variable (`price`), since it contains valid nightly price information for all listings.
