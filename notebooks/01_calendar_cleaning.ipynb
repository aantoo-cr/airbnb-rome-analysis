## 4. Data Source Selection

The calendar dataset was initially considered as the source of the target variable. However, `calendar.csv` does not include `price` or `adjusted_price` columns at all — it only contains `listing_id`, `date`, `available`, `minimum_nights`, and `maximum_nights`.

Therefore, the listings dataset was selected as the source of the target variable (`price`), since it contains valid nightly price information for all listings.

## 5. Conclusion

`calendar.csv` is a purely date-level availability table (`listing_id`, `date`, `available`,
`minimum_nights`, `maximum_nights`) — it has no `price` or `adjusted_price` column at all, so
there is nothing to impute or clean here on that front. Since `listings.csv` provides a valid,
fully-populated `price` per listing (see Section 8), that is the dataset used going forward as
the source of the target variable for the price-prediction models.
